```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

veriler = pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
N=10000 #toplam 10000 tiklama
d=10 #toplam 10 ilan var
#Ri(n)
oduller=[0] * d #ilk basta butun ilanlarin odulu
#Ni(n)
tiklamalar=[0] * d #o ana kadarki tiklamalar
toplam=0 #toplam odul
secilenler=[]

for n in range(0,N):

    ad=0 #secilen ilan
    max_ucb=0

    for i in range(1,d):
        if(tiklamalar[i]>0):
            ortalama=oduller[i] / tiklamalar[i]
            delta=math.sqrt(3/2*math.log(n)/tiklamalar[i])
            ucb=ortalama+delta
        else:
            ucb=N*10

        if max_ucb < ucb : #max'tan buyuk bir ucb cikti
            max_ucb=ucb
            ad=i

    secilenler.append(ad)
    tiklamalar[ad]+=1
    odul=veriler.values[n,ad]
    oduller[ad]+=odul
    toplam+=odul

print("Toplam Ödül")
print(toplam)
```

```python
plt.hist(secilenler)
plt.show()
```
